```python
week1 = [15,22,18,25,17,21]
week2 = [20,25,19,28,23,30]
```

```python
import statistics as st
```

```python
print(st.mean(week1))
print(st.median(week1))
print(st.mode(week1))
print(st.stdev(week1))
print(max(week1) - min(week1))
```

```python
print(st.mean(week2))
print(st.median(week2))
print(st.mode(week2))
print(st.stdev(week2))
print(max(week2) - min(week2))
```
